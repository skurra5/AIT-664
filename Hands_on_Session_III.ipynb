```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Data Preprocessing:

# The Iris dataset contains 150 samples of iris flowers, with 4 features and 3 target classes (species).
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a DataFrame with the data
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(y, iris.target_names)

# Check for missing values
print("Check for missing values:")
print(df.isna().sum())
```

```python
# Check normalization of numerical data using StandardScaler
print("Check normalization of numerical data using StandardScaler:")
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.iloc[:, :-1])
print(scaled_data[:5])
```

```python
# Display the preprocessed DataFrame
print("Preprocessed Data:")
print(df.head())
```

```python
# Group the data by species and calculate the mean of each feature
grouped_data = df.groupby('species').mean()

# Create a bar graph with the mean values for each species
plt.figure(figsize=(8,6))
grouped_data.plot(kind='bar', rot=0)

# Add title and labels to the graph
plt.title('Mean Values of Iris Features by Species')
plt.xlabel('Features')
plt.ylabel('Mean Values')

# Display the plot
plt.show()
```

```python
# Create a line chart with the data
plt.figure(figsize=(8,6))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.plot(subset.index, subset['sepal length (cm)'], label=species)

# Add title and labels to the graph
plt.title('Iris Features by Species')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')

# Display the legend and gridlines
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Line Chart 2
plt.figure(figsize=(10, 6))
for spec in df['species'].unique():
    subset = df[df['species'] == spec]
    plt.plot(subset.index, subset['petal length (cm)'], label=spec, marker='o', linestyle='-', markersize=8)
plt.title('Trend of Petal Length Over Samples')
plt.xlabel('Sample Index')
plt.ylabel('Petal Length (cm)')
plt.legend()
```

```python
# Create a scatter plot with the data
plt.figure(figsize=(8,6))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species)

# Add title and labels to the graph
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Display the legend and gridlines
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Scatter Plot 2
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='species', size='sepal length (cm)', sizes=(50, 200), palette='Set1')
plt.title('Petal Length vs Petal Width with Sepal Length Encoding')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
```

```python
# Add 'species' data to DataFrame
df['species'] = iris.target

# Convert 'species' to numerical labels
df['species'] = df['species'].astype('category')

# Create a heatmap to visualize the correlation between the features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)

# Add title and labels to the plot
plt.title('Heatmap of Feature Correlation')
plt.xlabel('Features')
plt.ylabel('Features')

# Display the plot
plt.show()
```

```python
# Calculate the frequency of each species
species_freq = df['species'].value_counts()

# Create a pie chart with the data
plt.figure(figsize=(8,6))
plt.pie(species_freq, labels=species_freq.index, autopct='%1.1f%%')

# Add title to the graph
plt.title('Frequency/Proportions of Species')

# Display the plot
plt.show()
```

```python
# Create a box plot to visualize the distribution of sepal lengths for each species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df)

# Add title and labels to the plot
plt.title('Distribution of Sepal Lengths for Each Species of Iris')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

# Display the plot
plt.show()

# Box Plot 2
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal length (cm)', data=df, palette='Set2')
plt.title('Distribution of Petal Length Across Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

plt.tight_layout()
plt.show()
```
